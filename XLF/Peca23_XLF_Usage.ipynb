# Using the Peca23_XLF.py Module
This notebook demonstrates how to use the `Peca23_XLF.py` script to compute or integrate the X-ray Luminosity Function from Peca et al. (2023).

Note: Luminosity is in log10 scale and in the 2-10 keV band.

```python
# If needed, install required packages
# !pip install numpy scipy matplotlib
```

```python
# Import the module
import Peca23_XLF as peca
import sys
sys.argv = ['Peca23_XLF.py', '44.5', '1.0', '--plot', 'no']
```

## Example 1: Evaluate space density at a single point of luminosity and redshift

```python
# log Lx = 44.5, z = 1.0
%run Peca23_XLF.py 44.5 1.0 --plot yes
```

## Example 2: Integrate over luminosity range at fixed redshift

```python
# log Lx = 43 to 45, z = 1.0
%run Peca23_XLF.py 43 45 1.0 --plot yes
```

## Example 3: Integrate over luminosity and redshift ranges

```python
# log Lx = 43 to 45, z = 0.5 to 2.0
%run Peca23_XLF.py 43 45 0.5 2.0 --plot yes
```
